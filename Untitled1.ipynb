```python
import os
import pandas as pd

df_dataset = pd.read_csv('../数据集/receipts3-002UP.csv',sep=';')
df_datasets = df_dataset.drop(df_dataset[df_dataset['Unnamed: 2']=='No'].index)
df_datasets=df_datasets.drop(columns=['Unnamed: 0','Unnamed: 2','Unnamed: 3'])
#df_datasets.reindex(range(900),method='nearest')
df_datasets['matched_bigram']='' 
df_datasets['matched_trigram']=''
df_datasets['matched_multi']=''
df_datasets
```

```python
def create_grams(text,n):
    list=[]
    gram=n
    string_grams = ""
    for i in range(0, len(text)-gram+1):
        x=text[i:i+gram]
        list.append(x)
        string_grams = ' '.join(list)
    return string_grams
def creat_bigrams(text):
    return create_grams(text,2)
def creat_trigrams(text):
    return create_grams(text,3)
```

```python
import os
import re
import pandas as pd
df_database = pd.read_csv('../yazio_all.csv')
regex = re.compile('[^a-zA-ZäöüÄÖÜß.-]')
df_database['name_lower']=df_database['item_name'].apply(lambda x:' '.join([regex.sub('', word).lower() for word in x.split()]))
df_database=df_database.drop(columns=['Unnamed: 0'])
df_database=df_database.where(df_database.notnull(),None)
df_database['bigram']=df_database['item_name'].apply(creat_bigrams)
df_database['trigram']=df_database['item_name'].apply(creat_trigrams)
df_database.head()
```

```python
df_database = df_database[['L1','L2','L3','label','item_name','name_lower','bigram','trigram','shop','nutrition']]
df_database.head()
```

```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
import json
es = Elasticsearch(['http://localhost:9200'],http_auth=['elastic', 'piPh2iel5aat'])
def delete_index() -> bool:
        """
        remove index
        :return: bool
        """
        try:
            es.indices.delete(index='es1', ignore=[400, 404])
            print(f"[INFO] Index 'es1' deleted successfully")
            return True
        except Exception as ex:
            print("[WARNING] some exception has occurred!")
            return False
def index_dataframe(index_name,df):
    delete_index()
    bulk_data = []
    df_dicts = df.to_dict(orient='records')
    for doc in df_dicts:
        action = {"_index": index_name, "_source": doc}
        bulk_data.append(action)
    try:
        helpers.bulk(es, bulk_data, stats_only=False)
        print("dataframe indexed!")
    except Exception as e:
        print(str(e))
```

```python
index_dataframe('es1',df_database)
```

```python
def search(item, search_status):
	search = json.dumps({})
	if search_status == "bigrams":
		search = json.dumps({
			'query': {
					'bool': {
						'should': [
							{
								'match': {
									'bigram': creat_bigrams(item),
								}
							},
						]
					}
				}
		})
	elif search_status == "trigrams":
		search = json.dumps({
					'query': {
						'bool': {
							'should': [
								{
									'match': {
										'trigram': creat_trigrams(item),
									}
								},
							]
						}
					}
				})
	elif search_status == "multi":
		search = json.dumps({
					'query': {
						'bool': {
							'should': [
								{'match': {'bigram': creat_bigrams(item)}},
                                {'match': {'trigram': creat_trigrams(item)}},
                                {'match': {'item_name': item}}
							]
						}
					}
				})
	else: 
		search = json.dumps({'query': {'bool': {'should': [{'match': {
										'item_name': item
									}}]}}})
	# to the job.
	result = es.search(index='es1', body=search, size=1)
	if not result['timed_out']:
		data = result['hits']['hits']
		if len(data) > 0:
			digital_items = []
			for item in data:
				digital_items.append(item['_source'])
			return digital_items
	return []
```

```python
def match_grams(text,n):
    list=[]
    string_grams = ""
    for i in search(text,n):
        x=i['item_name']+';'
        list.append(x)
        string_grams = ' '.join(list)
    return string_grams
def match_bigrams(text):
    return match_grams(text,"bigrams")
def match_trigrams(text):
    return match_grams(text,"trigrams")
def match_names(text):
    return match_grams(text,"")
def match_multi(text):
    return match_grams(text,"multi")
```
